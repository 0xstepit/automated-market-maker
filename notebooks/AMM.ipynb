```python
import numpy as np
import matplotlib.pyplot as plt
from helper.plot import make_plot, make_plot_price, make_plot_just_line
from helper.models import *
```

```python
# Equilibrium condition
amount_x = 5
amount_y = 5
D_sum = amount_x + amount_y
D_prod = amount_x * amount_y

# Considered token X liquidity
x = np.linspace(0.01, 20, int(20/0.01))
```

```python
y_pi = compute_y_product_invariant(D_prod, x)
points_ = np.array([2.5, 7.5])
y_pi_points = compute_y_product_invariant(D_prod, points_)
data = [[x, y_pi]]
points = [points_, y_pi_points]
legend = []
title = 'Constant Product Invariant'
make_plot_price(data, points, title, 20, legend, save_fig=True, fig_title='xyk.png')
```

```python
y_li = compute_y_linear_invariant(D_sum, x)
y_pi = compute_y_product_invariant(D_prod, x)
y_sw = compute_y_stableswap_no_chi(D_sum, x)

data = [[x, y_li], [x, y_pi], [x, y_sw]]
legend = ['Linear invariant', 'Product invariant', 'Stableswap']
title = 'Automated Market Maker'
make_plot(data, title, 20, legend, save_fig=True, fig_title='amm.png')
```

```python
legend = []
data = []
for chi in range(1, 11):
    data.append([x, compute_y_stableswap_static(D_sum, x, 2 ** chi)])
    legend.append(r'$\chi = $'+ f'{2 ** chi}')

title = 'Stableswap with static leverage parameter'
make_plot(data, title, 20, legend, save_fig=True, fig_title='stableswap-static.png')
```

```python
# This plot is made just for my personal landing page https://stepyt.github.io/
make_plot_just_line(data, title, 20, legend, save_fig=True, fig_title='stableswap-static-transparent.png')
```

```python
legend = []
data = []
A_list = [0.1, 3, 100]
for A in A_list:
    data.append([x, compute_y_stableswap_dynamic(D_sum, x, A)])
    legend.append(r'A = '+ f'{A}')

title = 'Stableswap with dynamic leverage parameter'
make_plot(data, title, 20, legend, save_fig=True, fig_title='stableswap-dynamic.png')
```
